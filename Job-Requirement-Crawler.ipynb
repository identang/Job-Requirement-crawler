```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup

def parse_elements(url, element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    return elements

# A page with multiple Linkedin job postings
url = 'https://www.linkedin.com/jobs/search/?currentJobId=3679095269&keywords=frontend%20web%20developer'

# Parse the anchor elements
link_details = parse_elements(url, 'a')

job_list = []

# Get the links from the job posts
for link in link_details:
    if 'href' in link.attrs and 'class' in link.attrs:
        link_classes = link.attrs['class']
        if 'base-card__full-link' in link_classes:
            job_url = link.attrs['href']
            job_list.append(job_url)


print(job_list)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random

# Put all the description in one string
job_requirements = ""
driver = webdriver.Chrome()

# Choose 3 links to avoid AuthBlock from Linkedin, can be solved if scraping with multiple IPs
for job in job_list[4:7]: 
    driver.get(job)


   # Wait for the <article> tag with class 'jobs-description__container' to be present
    wait = WebDriverWait(driver, 10)
    div_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'show-more-less-html__markup')))
        
    print(div_element.text)
    div_html = div_element.get_attribute('innerHTML')
        
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(div_html, 'html.parser')
        
    # Find all <li> elements within the <div> element
    li_elements = soup.find_all('li')
    
    # Extract and print the text content of each <li> element
    for li in li_elements:
        job_requirements += li.get_text(strip=True) + '\n'
    # Avoid AuthBlock from Linkedin by simulating human browsing
    delay = random.uniform(1,3)
    time.sleep(delay)    

driver.quit()
```

```python
# Remove words without insight
words_to_remove = ['Front End', 'End', 'Front', 'development','ability', 'experience','work','create','working','IT','years','Web', 'Ability']
for word in words_to_remove:
    job_requirements = job_requirements.replace(word, '')
    
print(job_requirements)
```

```python
# Import NLP library to extract the keywords or key phrases of the job
import yake
kw_extractor = yake.KeywordExtractor()

language = "en"
max_ngram_size = 10 # The maximum size of the extracted keywords
deduplication_threshold = 0.5 # value for controlling the duplication of keywords
numOfKeywords = 30 # Numbers of keywords generated
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(job_requirements)
for kw in keywords:
    print(kw)
```
